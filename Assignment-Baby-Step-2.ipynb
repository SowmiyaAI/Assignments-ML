```python
# The pandas is a libraries it has globalised for the all data base related function. 
# It's helps to table handling and file reading purpose. 
# The as - It's helps to converting like pandas convert to pd
import pandas as pd
```

```python
# How to Load the dataset.
# Call the read_csv function in class pd. 
# Pass the dataset csv file in the ().
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# Hoe to view the dataset
dataset
```

```python
# Assign the particular column(input) in variable name from whole dataset.
# How to view the assigned varialble
indepentent=dataset[["Years _Of _Experience"]]
indepentent
```

```python
# Assign the particular column(output) in variable name from whole dataset.
# How to View the assigned varialble
depentent=dataset[["Salary"]]
depentent
```

```python
# How to Split trainset and testset
from sklearn.model_selection import train_test_split
train_test_split(indepentent,depentent,test_size=0.30,random_state=0)
```

```python
# How to Split trainset and testset
# Import the function train_test_split from sklearn.model_selection.
# Sklearn support to machine learning. And The sklearn.model_selection used to split train set and test set.
# Pass the parameters to 4 variables(train set input, test set input, train set output, test set output).
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indepentent,depentent,test_size=0.30,random_state=0)
```

```python
# How to Create the Model
# Import the function LinearRegression from sklearn.linear_model.
# Non parameterised function(LinearRegression) assigned to variable. It's consider as a class 
# Call the fit function using access dot operator and pass the train set input and output
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# How to View the slope
# The weight value is saved in regressor variable.
# Weight is calculate in coef_ methoed.
# regressor.coef_ value is assigned in variable.
weight=regressor.coef_
weight
```

```python
# How to View the bais
# The bais vlue is saved in regressor variable.
# Bais is calculate in intercept_ methoed
# regressor.intercept_ value is assigned in variable.
bais=regressor.intercept_
bais
```

```python
# How to Evaluate
# The model is saved in regressor variable.
# Pass the testset input value to predict function.
# regressor.predict(x_test) value is saved in the y_pred variable.
y_pred=regressor.predict(x_test)
```

```python
# How to find the r2
# Import the function r2_score from sklearn.metrics
# Pass the actual output(y_test) and predicted value to r2_score function.
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# How to view the r2_score value
# r2_score vale nearly 1 it is a good model
# r2_score vale nearly 0 it is a poor model
r_score
```

```python
# How to create the model name.
# Import the pickle libraries
# Give the name to model(Filename_Algorithamname.sav((.sav is Pickle extension)
import pickle
filename="finalized_model_linear.sav"
```

```python
# How to save the model
# Pass the model in the filename(finalized_model_linear.sav)
pickle.dump(regressor,open(filename,'wb'))
```

```python
# How to load and re-check the model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
